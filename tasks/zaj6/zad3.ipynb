Załaduj dane z pliku "dane-errors.csv" plik ten zawiera kolejne pomiary Berylu 7 w atmosferze. Tym razem zawiera on dodatakowo niepewność pomiarową owych pomiarów. 

Zadanie polega na policzeniu średnich miesięcznych w sposób "poprawny", tj. średnią trzeba ważyć za pomocą niepewności pomiarowych. Proszę zastosować funkcję gropupby

```python
import pandas as pd
data = pd.read_csv('data-errors.csv', header=2)

from datetime import date, timedelta
def week_start_date(year, week, dow=0):    
    """
    :param int year: Numer roku
    :param int week: Numer tygodnia w roku 
    :param int dow: Funkcja ta zwraca datę, zwróci datę dnia o indeksie dow, zero oznacza poniedziałek, jeden wtorek itp. 
    """
    # http://stackoverflow.com/a/1287862/7918
    d = date(year, 1, 1)    
    delta_days = d.isoweekday() - 1
    delta_weeks = week
    if year == d.isocalendar()[0]:
        delta_weeks -= 1
    delta = timedelta(days=-delta_days+dow, weeks=delta_weeks)
    return d + delta

dates = []
for y,w in zip(data['Rok'],data['Tydzień']):
    dates.append(week_start_date(int(y),int(w),0))
data['Date'] = dates
```
